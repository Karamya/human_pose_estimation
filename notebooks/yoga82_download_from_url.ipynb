```python
url = 'https://tinypng.com/images/social/website.jpg'

def testRequest():
    image_name = 'test1.jpg'
    r = requests.get(url, stream=True)
    with open(image_name, 'wb') as f:
        for chunk in r.iter_content():
            f.write(chunk)

def testRequest2():
    image_name = 'test2.jpg'
    r = requests.get(url)
    i = Image.open(BytesIO(r.content))
    i.save(image_name)
    
def testRequest3():
    image_name = 'test2b.jpg'
    r = requests.get(url)
    with open(image_name, 'wb') as f:
        f.write(r.content)

def testUrllib():
    image_name = 'test3.jpg'
    urllib.request.urlretrieve(url, image_name)

def testwget():
    image_name = 'test4.jpg'
    wget.download(url, image_name)
```

```python
def download(line, sess, default_folder = Path("../Yoga-82/")):
    dest, src = line.split("\t")
    dest = default_folder / dest
    dest.parent.mkdir(parents=True, exist_ok=True)
    r = sess.get(src)
    with open(dest, 'wb') as outfile:
        outfile.write(r.content)
        
def download2(line, default_folder = Path("../Yoga-82/")):
    
    dest, src = line.split("\t")
    dest = default_folder / dest
    dest.parent.mkdir(parents=True, exist_ok=True)
    os.system(f"wget --tries=5 -O {str(dest)} {str(src)}")
    
def download3(line, default_folder = Path("../Yoga-82/")):
    
    dest, src = line.split("\t")
    dest = default_folder / dest
    dest.parent.mkdir(parents=True, exist_ok=True)
    wget.download(str(src), str(dest))
    
    
def requestsdownload(line, default_folder = Path("../Yoga-82/")):
    dest, src = line.split("\t")
    dest = default_folder / dest
    dest.parent.mkdir(parents=True, exist_ok=True)
    
    with requests.get(src, stream=True) as r:
        with open(dest, "wb") as f:
            r.raw.decode_content = True
            shutil.copyfileobj(r.raw, f)
```

```python
asana_file = '../Yoga-82/yoga_dataset_links/Upward_Bow_(Wheel)_Pose_or_Urdhva_Dhanurasana_.txt'
with open(asana_file, 'r') as f:
    session = requests.Session()
    retry = Retry(connect=3, backoff_factor=0.5)
    adapter = HTTPAdapter(max_retries=retry)
    session.mount('http://', adapter)
    session.mount('https://', adapter)
    for line in f:
        try:
            download(line, session)
        except urllib.error.HTTPError as err:
            if err.code == 404:
                print(line.split('\t')[1], " is ignored due to 404 error")
            elif err.code == 403:
                download(line, session)
            elif err.code == 400:
                pass
            else:
                
                raise err
        except:
            print(line.split('\t')[1], " is ignored")
```
