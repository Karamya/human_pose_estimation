```python
def download(line, sess, default_folder = Path("../Yoga-82/")):
    dest, src = line.split("\t")
    dest = default_folder / dest
    dest.parent.mkdir(parents=True, exist_ok=True)
    r = sess.get(src)
    with open(dest, 'wb') as outfile:
        outfile.write(r.content)
        
def download2(line, default_folder = Path("../Yoga-82/")):
    
    dest, src = line.split("\t")
    dest = default_folder / dest
    dest.parent.mkdir(parents=True, exist_ok=True)
    os.system(f"wget --tries=5 -O {str(dest)} {str(src)}")
    
def download3(line, default_folder = Path("../Yoga-82/")):
    
    dest, src = line.split("\t")
    dest = default_folder / dest
    dest.parent.mkdir(parents=True, exist_ok=True)
    wget.download(str(src), str(dest))
    
    
def requestsdownload(line, default_folder = Path("../Yoga-82/")):
    dest, src = line.split("\t")
    dest = default_folder / dest
    dest.parent.mkdir(parents=True, exist_ok=True)
    
    with requests.get(src, stream=True) as r:
        with open(dest, "wb") as f:
            r.raw.decode_content = True
            shutil.copyfileobj(r.raw, f)
```

![Human Pose Estimation](../pose_tracking_full_body_landmarks.png)

```python
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
```

```python
def detect_pose(filepath):
    with mp_pose.Pose(static_image_mode=True, model_complexity=2, min_detection_confidence=0.5) as pose:
        image = cv2.imread(filepath)
        
        image_height, image_width, _ = image.shape
        
        # Convert the BGR image to RGB before processing.
        results = pose.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        
        if not results.pose_landmarks:
            annotated_image = image.copy()
            
        # Draw pose landmarks on the image.
        annotated_image = image.copy()
        mp_drawing.draw_landmarks(
            annotated_image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
        annotated_image = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)
    return results, plt.imshow(annotated_image)
```

```python
def copy_detectable_images(img_path):
    with mp_pose.Pose(static_image_mode=True, model_complexity=2, min_detection_confidence=0.5) as pose:
        image = cv2.imread(str(img_path))
        
        if image is not None:
            # Convert the BGR image to RGB before processing.
            results = pose.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

            if results.pose_landmarks:
                dest = Path(str(img_path).replace("Yoga-82", "Yoga-82_train"))
                os.makedirs(dest.parent, exist_ok=True)
                shutil.copy(img_path, dest)
            else:
                print(f"Ignored copying {img_path}")
        else:
            print("opencv could not read ", img_path)
    return
```

```python
for img_file in Path('../Yoga-82/').rglob('*.jpg'):
    copy_detectable_images(img_file)
```

## Create test data having a train:test ratio of 70:30

```python
for folder in Path("../Yoga-82_train/").glob("*"):
    filepaths = list(folder.glob("*.jpg"))
    test_size = int(len(filepaths)*0.3)
    test_files = np.random.choice(filepaths, size=test_size, replace=False)
    os.makedirs(str(filepaths[0].parent).replace("Yoga-82_train", "Yoga-82_test"), exist_ok=True)
    for test_file in test_files:
        dest = str(test_file).replace("Yoga-82_train", "Yoga-82_test")
        shutil.move(test_file, dest)
```
